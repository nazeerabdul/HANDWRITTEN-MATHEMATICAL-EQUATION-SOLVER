```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_train=pd.read_csv("train_final3.csv",index_col=False)
df_train.head()
```

```python
labels=df_train[['784']]
labels
```

```python
df_train.drop(df_train.columns[[784]],axis=1,inplace=True)
df_train.head()
```

```python
np.random.seed(1212)
import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
from keras.layers import Input,Dense
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
```

```python
labels=np.array(labels)
len(labels)
labels
```

```python
df_train
```

```python
from keras.utils.np_utils import to_categorical
cat=to_categorical(labels,num_classes=14)
```

```python
df_train.head()
```

```python
l=[]
for i in range(67453):
    l.append(np.array(df_train[i:i+1]).reshape(28,28,1))
```

```python
len(l)
```

```python
np.random.seed(7)
```

```python
model=Sequential()
model.add(Conv2D(30,(5,5),input_shape=(28,28,1),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(15,(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(14,activation='softmax'))
```

```python
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(np.array(l),cat,shuffle=True,epochs=10)
```

```python
def printEq(x):
    flag=0
    for i in range(len(x)):
        if flag==1:
            flag=0
            continue
        if i!=len(x)-1 and x[i]=='X'and x[i+1]!='-' and x[i+1]!='+' and x[i+1]!='*':
            print('X^'+str(x[i+1]), end=" ")
            flag=1
        else :
            print(x[i], end=" ")
```

```python
def highdegree(x):
    Max=1
    for i in range(len(x)):
        if i!=len(x)-1 and x[i]=='X' and x[i+1]!='-' and x[i+1]!='+' and x[i+1]!='*':
            Max=max(Max,x[i+1])
    return Max  
```

```python
def solution(x):
    printEq(x)
    print("")
    degree=highdegree(x)
    eq=np.ones(degree+1)
    ind=0
    for k in range(degree+1):
        #find constant term
        if k==0 :
            for i in range(len(x)):
                #x not infront , x not in back
                if i==0:
                    if x[i+1]!='X' and int(x[i])<10:
                        eq[degree-k]=x[i]
                        break
                if i==len(x)-1:
                    if x[i-1]=='-':
                        eq[degree-k]=float(x[i])*-1
                        break
                    else:
                        eq[degree-k]=x[i]
                        break
                if x[i+1]!='X' and x[i-1]!='X' and x[i].isdigit():
                    if x[i-1]=='-':
                        eq[degree-k]=float(x[i])*-1
                        break
                    else:
                        eq[degree-k]=x[i]
                        break
        if k==1:
            for i in range(len(x)):
                if i==len(x)-2:
                    eq[degree-k]=x[i]
                if(i<len(x)-2):    
                    if x[i+1]=='X' and x[i].isdigit():
                        if(not (x[i+2].isdigit())):
                            if x[i-1]=='-':
                                eq[degree-k]=float(x[i])*-1
                                break
                            else:
                                eq[degree-k]=x[i] 
                                break
                else:
                    eq[degree-k]=x[i-1]      
        if k>1:
            for i in range(2,len(x)):
                if int(x[i])==k and x[i-1]=='X':
                    eq[degree-k]=x[i-2]
                    break
    #Print and return ans
    print(eq)
    print(np.roots(eq))
    return np.roots(eq)
```
